```python
def countStudents(self, students, sandwiches):
    count = [0,0]
    for s in students:
        count[s] += 1   # populate count

    for sandwich in sandwiches:
        if not count[sandwich]: break   # if no one wants the top sandwich, break out of the loop
        if count[sandwich]:
            count[sandwich] -= 1

    return count[0] + count[1]
```

### 2. Peer Review & Draft Code

After looking at some of the top solutions on LeetCode, I realized almost all use the "Dutch National Flag" algorithm. It is used for sorting arrays with three distinct values at a time complexity of O(n). However, since
1. I dont know the algorithm (I dont want to be copying people's code)
2. The constraints say the max length of nums is 300, which is in the acceptable range for O(n^2) algorithms.

I will be using bubble sort.

```python
def sortColors(self, nums):
    n = len(nums)
    swapped = True
    while swapped:  # have we swapped any numbers when passing through the list?
        swapped = False
        i = 1
        while i < n:    # stops if index goes out of bound
            if nums[i-1] > nums[i]: # if left number is bigger than the right, we swap
                nums[i-1], nums[i] = nums[i], nums[i-1]
                swapped = True
            i +=1
```
