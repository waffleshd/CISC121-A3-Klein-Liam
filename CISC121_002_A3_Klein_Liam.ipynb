```python
'''
After reviewing this solution on LeetCode: https://leetcode.com/problems/valid-parentheses/solutions/7108838/100-beats-0ms-stack-beginner-friendly-py-y4xp/
I realized my approach was flawed in quite a fundamental way:
It could only handle nested brackets, such as ([{}]), but once a string like
()[]{} was inputted, it would return False, since it would compare ( and }.

Therefore I implemented the logic of that solution:
'''

def isValid(self, s):
    parenth_hash = {"(":")","[":"]","{":"}"}
    stack = []
    for char in s:
        if char in parenth_hash:    # if the parenthesis is ( [ or {
            stack.append(char)
        # if not, we check if the latest addition to the stack is the complement to the closed parenthesis
        elif char == parenth_hash[stack.pop()]:
            continue
        else: return False  # if none of these apply, then we know that there is an incorrect set of parentheses
    return True

```
